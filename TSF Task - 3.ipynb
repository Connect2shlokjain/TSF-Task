Using HeatMap to understand various Properties with diffrent graphs below

```python
print("Total Sales:",df['Sales'].sum())
print("Total Profit:",df['Profit'].sum())
sns.heatmap(df_corr,annot=True)
```

```python
df['Region'].value_counts().plot.pie(colors=['pink','lightgreen','skyblue','lightyellow'],autopct="%.1f%%")
```

```python
df.groupby(['Region'])['Sales','Profit'].sum().plot.bar()
plt.xlabel("REGION")
plt.title("REGION WISE SALES AND PROFIT")
```

```python
sns.countplot(x=df['Ship Mode'])
```

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show
```

```python
plt.figure(figsize=(14,9))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show
```

```python
top_state=df.groupby('State').sum().sort_values('Profit',ascending=False)
top_state.head(16)
```

```python
#Calculating Cost and Profit in Percentage
df['Cost']=df['Sales']-df['Profit']
df['Cost'].head()
```

```python
#Calculating Profit Percentage
df['Profit %']=(df['Profit']/df['Cost'])*100
```

Summaries The Given Data and making a table out of them in Highest to Lowest rate

```python
DataFrame=df.sort_values(['Profit %','Sub-Category'],ascending=False).groupby('Profit %').head(11)
DataFrame
```
